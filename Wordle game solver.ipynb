> Execute all the code in the current order

> For the final block enter the word to be guessed and the first guess

> The code will return a list of all the guesses that the computer performed

```python
import pandas as pd
import numpy as np

# df = pd.read_csv('words_alpha.txt',delimiter='/n',engine='python')
df = pd.read_csv('mitwordlist.txt.txt',delimiter='/n',engine='python')
dc = df.to_numpy()
dc
```

```python
flw = []

# flw.append('xyz')

counter = -1

for i in dc:
    if (type(i[0]) != str):
        continue
    if(len(i[0])==5):
        flw.append(i[0])
flw
        
```

```python
def chkWrds(g,w):
    retVal = [['_',''],['_',''],['_',''],['_',''],['_','']]
    wrongLet = []
    combArr = []
    for posWrd in range(0,5):
        inWrd = False
        let = g[posWrd]
#         print(let)
        if (let in w):
#             print(w.index(let))
            retVal[posWrd][1]=let
            if(posWrd==w.index(let)):
                retVal[posWrd][0]='*'
            else:
                retVal[posWrd][0]='?'
        else:
            wrongLet.append(let)
    if(retVal[0][0]==retVal[1][0]==retVal[2][0]==retVal[3][0]==retVal[4][0]=='*'):
        return g
    combArr.append(retVal)
    combArr.append(wrongLet)
    return combArr
```

```python
def newList(chkWrdResult, prevFinalArr):
    retVal = chkWrdResult[0]
#     print(retVal)
    wrongLet = chkWrdResult[1]
#     print(wrongLet)
    finalArr = []
    
    for wrd in prevFinalArr:
        wrongWrd = False
        
        for wLet in wrongLet:
            if(wLet in wrd):
                wrongWrd = True
                break
            
        if(wrongWrd):
            continue
        
        for val in range(0,5):
            let = retVal[val]
            if(let[0]=='_'):
                continue
            elif(let[1] in wrd):
                if(let[0]=='?'):
                    if(val==wrd.index(let[1])):
                        wrongWrd=True
                elif(let[0]=='*'):
                    if(val!=wrd.index(let[1])):
                        wrongWrd=True
            else:
                wrongWrd=True
            
            if(wrongWrd):
                break
        
        if(wrongWrd):
            continue
        else:
            finalArr.append(wrd)
    return finalArr
```

```python
def guessMostLikely(chkWrdsOtp, finalList):
    if (len(chkWrdsOtp)==1):
        return chkWrdsOtp[0]
    
    checkWrds = chkWrdsOtp[0]
    indexNum = 0
    mostLetCounter = 0
    curLetCounter = 0
    curLet = ''
    curGWord = ''
    for index in range(0,5):
        spot = checkWrds[index]
        if(spot[0]=='_'):
            indexNum = index
    
    for i in range(len(finalList)):
        wrd = finalList[i]
        indLet = wrd[indexNum]
        
        if(indLet!=curLet):
            if(curLetCounter>mostLetCounter):
                curLet==indLet
                mostLetCounter=curLetCounter  
                curLetCounter=0
                curGWord = wrd 
        curLetCounter= curLetCounter+1
    
    if(curLetCounter>mostLetCounter):
        curLet==indLet
        mostLetCounter=curLetCounter  
        curLetCounter=0
        curGWord = wrd
            
    return curGWord                   
```

```python
def wholeProcess(fGuess,rWord):
    
    
    wordGuessed = False
    curList = []
    guessList = []
    guessList.append(fGuess)
    guess = fGuess
    if(len(guessList)==1):
        curList = flw
    while(wordGuessed==False):
        otp1 = chkWrds(guess,rWord)

        if(otp1==rWord):
            wordGuessed = True
        if(wordGuessed):
            return guessList

        otp2 = newList(otp1,curList)
        curList=otp2

        guess = guessMostLikely(otp1,otp2)

        guessList.append(guess)
        
        if(guess==rWord):
            wordGuessed = True
        if(wordGuessed):
            return guessList
    
```

```python
realWord = input("Real Word: ")
firstGuess = input("First Guess: ")
print(wholeProcess(firstGuess,realWord))
```
